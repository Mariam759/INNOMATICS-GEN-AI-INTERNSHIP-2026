### MOBILE RECHARGE VALIDATION 

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    
    if amount < 50:
        print(" Recharge amount must be at least â‚¹50.")
        return False
    elif amount not in valid_plans:
        print(" Invalid plan. Choose from 199, 299, 399, 599.")
        return False
    else:
        print("Recharge successful!")
        return True


# Retry system using while loop
while True:
    try:
        amount = int(input("Enter recharge amount: â‚¹"))
        
        if validate_recharge(amount):
            break  # Exit loop if valid
        else:
            print("Please try again.\n")
            
    except ValueError:
        print("Invalid input! Please enter a number.\n")
```

### INVENTORY REORDER ALERT SYSTEM

```python
def check_inventory(stock_dict):
    for product, quantity in stock_dict.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert (Stock = {quantity})")
        else:
            print(f"{product}: Stock OK (Stock = {quantity})")


# Sample dictionary
inventory = {
    "Rice": 10,
    "Wheat": 25,
    "Sugar": 12,
    "Oil": 30
}

# Function call
check_inventory(inventory)
```

### Student result processing system 

```python
def process_result(marks):
    total = 0
    
    # Calculate total using loop
    for mark in marks:
        total += mark
    
    average = total / len(marks)
    
    # Decide result
    if average >= 50:
        return average, "Pass"
    else:
        return average, "Fail"


# Input marks
marks = list(map(int, input("Enter marks separated by space: ").split()))

# Function call
avg, result = process_result(marks)

print(f"Average Marks: {avg:.2f}")
print(f"Result: {result}")
```

### Cab fare estimator

```python
def calculate_fare(distance, is_peak):
    base_fare = 50
    fare = base_fare + (12 * distance)
    
    if is_peak:
        fare += fare * 0.25   # 25% extra
    
    return fare


# Retry system
while True:
    try:
        distance = float(input("Enter distance (in km): "))
        peak = input("Is it peak hour? (yes/no): ").lower()
        
        is_peak = True if peak == "yes" else False
        
        fare = calculate_fare(distance, is_peak)
        
        print(f" Total Fare: â‚¹{fare:.2f}")
        break   # exit loop after success
    
    except ValueError:
        print(" Invalid input! Please enter correct values.\n")
        print(" Please try again.\n")
```

### employee attendance estimation 

```python
def check_eligibility(attendance):
    present_count = 0
    
    # Count "P" using loop
    for day in attendance:
        if day.upper() == "P":
            present_count += 1
    
    total_days = len(attendance)
    percentage = (present_count / total_days) * 100
    
    if percentage >= 75:
        return percentage, "Eligible"
    else:
        return percentage, "Not Eligible"


# Input attendance
attendance = input("Enter attendance (P/A separated by space): ").split()

# Function call
percent, result = check_eligibility(attendance)

print(f"Attendance Percentage: {percent:.2f}%")
print(f"Status: {result}")
```

### Password strength checker

```python
def check_password_strength(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"
    
    # Check each character using loop
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True
    
    # Conditions
    if len(password) < 8:
        return " Weak Password (Minimum 8 characters required)"
    elif not has_digit:
        return " Weak Password (At least one digit required)"
    elif not has_special:
        return " Weak Password (At least one special character @#$ required)"
    else:
        return " Strong Password"


# Input
password = input("Enter your password: ")

# Function call
result = check_password_strength(password)

print(result)
```
