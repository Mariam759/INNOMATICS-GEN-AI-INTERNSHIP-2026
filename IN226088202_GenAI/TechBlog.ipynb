```python
from langchain_groq import ChatGroq

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0
)

response = llm.invoke("Explain LangChain in simple terms")
print(response.content)
```

```python
from langchain_core.prompts import PromptTemplate

template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms"
)

print(template.format(topic="Machine Learning"))
```

```python
from langchain_core.prompts import ChatPromptTemplate

chat_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a Data Science expert"),
    ("user", "Explain {topic} with examples")
])

chain = chat_prompt | llm

response = chain.invoke({"topic": "Overfitting"})
print(response.content)
```

```python
chain = template | llm

response = chain.invoke({"topic": "Neural Networks"})
print(response.content)
```

```python
from langchain_core.messages import HumanMessage, AIMessage
from langchain_core.prompts import MessagesPlaceholder

chat_history = []

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant"),
    MessagesPlaceholder(variable_name="chat_history"),
    ("human", "{input}")
])

chain = prompt | llm

def chat(user_input):
    chat_history.append(HumanMessage(content=user_input))

    response = chain.invoke({
        "input": user_input,
        "chat_history": chat_history
    })

    chat_history.append(AIMessage(content=response.content))
    return response.content

print(chat("Hi, my name is Mariam"))
print(chat("What is AI?"))
print(chat("What is my name?"))
```

```python
from langchain_core.output_parsers import StrOutputParser

chain = template | llm | StrOutputParser()

result = chain.invoke({"topic": "Deep Learning"})
print(result)
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from sklearn.metrics.pairwise import cosine_similarity

embeddings = GoogleGenerativeAIEmbeddings(
    model="models/gemini-embedding-001"
)

vec1 = embeddings.embed_query("cat")
vec2 = embeddings.embed_query("dog")

similarity = cosine_similarity([vec1], [vec2])
print("Similarity:", similarity)
```

```python
from langchain_community.vectorstores import FAISS
from langchain_core.documents import Document

docs = [
    Document(page_content="LangChain helps build LLM apps"),
    Document(page_content="FAISS is used for vector search"),
]

vector_store = FAISS.from_documents(docs, embeddings)

print("Vectors stored:", vector_store.index.ntotal)
```

```python
from langchain_core.runnables import RunnablePassthrough

retriever = vector_store.as_retriever()

prompt = ChatPromptTemplate.from_template("""
Answer using context only:

{context}

Question: {question}
""")

def format_docs(docs):
    return "\n".join(doc.page_content for doc in docs)

rag_chain = (
    {
        "context": retriever | format_docs,
        "question": RunnablePassthrough()
    }
    | prompt
    | llm
    | StrOutputParser()
)

print(rag_chain.invoke("What is FAISS?"))
```

```python
from langchain_core.tools import tool
from langchain.agents import create_agent

@tool
def add_numbers(a: int, b: int) -> int:
    """Adds two numbers"""
    return a + b

tools = [add_numbers]

agent = create_agent(
    model=llm,
    tools=tools
)

result = agent.invoke({
    "messages": [("human", "Add 10 and 20")]
})

print(result["messages"][-1].content)
```
