```python
def smart_parking_v2(capacity, logs):
    total_in = 0
    total_out = 0
    current = 0
    peak = 0

    for log in logs:
        if log == "IN":
            total_in += 1
            current += 1
        elif log == "OUT":
            total_out += 1
            current -= 1

        peak = max(peak, current)

    current_vehicles = total_in - total_out

    print("Currently Parked Vehicles:", current_vehicles)
    print("Peak Parking Usage:", peak)

    if current_vehicles > capacity:
        print("ALERT: Parking Exceeded Capacity!")
    else:
        print("Parking Status: Available")


# CALL FUNCTION (this was missing)
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_v2(capacity, logs)
```

```python
def estimate_delivery_time(distance, traffic, weather):
    # Base average speed (km/h)
    speed = 25  

    # Base delivery time (in minutes)
    base_time = (distance / speed) * 60

    # Traffic delay (in minutes)
    traffic_delay = {
        "Low": 5,
        "Medium": 10,
        "High": 20
    }

    # Weather delay (in minutes)
    weather_delay = {
        "Clear": 0,
        "Rainy": 10,
        "Storm": 20
    }

    # Fixed handling delay (packing, pickup, etc.)
    handling_delay = 5

    # Total estimated time
    total_time = base_time \
                 + traffic_delay.get(traffic, 0) \
                 + weather_delay.get(weather, 0) \
                 + handling_delay

    print("Estimated Delivery Time:", int(total_time), "minutes")


# Sample Input
estimate_delivery_time(8, "High", "Rainy")
```

```python
def analyze_occupancy(total_seats, booked_seats):
    # count booked seats
    booked_count = len(booked_seats)

    # calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", int(occupancy), "%")

    # determine show status
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")

    # suggestion
    if occupancy >= 90:
        print("Suggestion: Open Additional Show")

total_seats = 200
booked_seats = [1] * 150   # 150 seats booked

analyze_occupancy(total_seats, booked_seats)
```

```python
def classify_server_load(cpu_readings):
    # calculate average CPU usage
    avg_cpu = sum(cpu_readings) / len(cpu_readings)

    print("Average CPU Load:", int(avg_cpu), "%")

    # classify server status
    if avg_cpu < 50:
        print("Server Status: Normal")
    elif 50 <= avg_cpu <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

classify_server_load(cpu_readings)
```

```python
def monitor_resources(resource_usage):
    overused = []

    # define usage threshold (in hours)
    threshold = 8

    # check each resource
    for resource, hours in resource_usage.items():
        if hours > threshold:
            overused.append(resource)

    # output results
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resources(usage)
```

```python
def manage_registrations(capacity, registrations):
    confirmed = 0
    waitlisted = 0

    # process registrations one by one
    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    # registration status
    if confirmed >= capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")


# Sample Input
manage_registrations(100, 105)
```
