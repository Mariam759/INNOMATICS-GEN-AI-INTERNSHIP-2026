```python
def analyze_engagement(likes_list):
    total_likes = 0
    
    for like in likes_list:
        total_likes += like
    
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Example
analyze_engagement([200, 300, 150, 400])
```

```python
def check_medicine_stock(stock):
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"
    
    print("Medicine Stock:", stock)
    print("Status:", status)


# Example
check_medicine_stock(6)
```

```python
def check_rainfall(rainfall_data, required_level):
    total = 0
    
    for rain in rainfall_data:
        total += rain
    
    average = total / len(rainfall_data)
    
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    print("Average Rainfall:", average)
    print("Rainfall Status:", status)


# Example
check_rainfall([70, 75, 68, 80, 67], 70)
```

```python
def detect_duplicates(usernames):
    unique_users = set(usernames)
    
    if len(unique_users) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example
detect_duplicates(["john", "alice", "john", "mark"])
```

```python
def check_eligibility(age):
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"
    
    print("Patient Age:", age)
    print("Eligibility Status:", status)


# Example
check_eligibility(21)
```

```python
def filter_premium_crops(prices):
    premium = []
    
    for price in prices:
        if price > 2000:
            premium.append(price)
    
    print("Premium Crops:", premium)


# Example
filter_premium_crops([1500, 2500, 1800, 3200])
```

```python
def check_system_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    
    print("Error Count:", errors)
    print("System Status:", status)


# Example
check_system_health(7)
```

```python
def check_transaction(amount):
    limit = 50000
    
    if amount <= limit:
        status = "Approved"
    else:
        status = "Rejected"
    
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)


# Example
check_transaction(60000)
```

```python
def check_attendance(attendance_list):
    total_classes = len(attendance_list)
    present_count = 0
    
    for status in attendance_list:
        if status == "P":
            present_count += 1
    
    percentage = (present_count / total_classes) * 100
    
    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"
    
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", eligibility)


# Example
check_attendance(["P", "P", "A", "P", "P"])
```

```python
def calculate_electricity_bill(units):
    bill = 0
    
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7
    
    # Usage Classification
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
    
    print("Total Bill:", bill)
    print("Usage Level:", usage)


# Example
calculate_electricity_bill(250)
```
