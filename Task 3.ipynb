### Problem Statement 1: Employee Performance Bonus Eligibility

```python
employees = {
    "Ravu": 92,
    "Anitha": 88,
    "Kiran": 92,
    "Suresh": 85
}

highest_score = max(employees.values())

top_performers =[]

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

print("Top Perforemrs Eligible for Bonus : ",
      ",".join(top_performers),
      f"(Score: {highest_score})")
```

### Problem Statement 2: Search Query Keyword Analysis

```python
query = input("enter your search query: ")

query = query.lower()

import string 
query = query.translate(str.maketrans("","",string.punctuation))

words = query.split()

frequency ={}

for word in words:
    frequency [word]= frequency.get(word, 0)+1

result={}

for word, count in frequency.items():
    if count >1:
        result [word] = count

print (result)
```

### Problem Statement 3: Sensor Data Validation

```python
user_input = input ("Enter sensor data seperated by space :")
sensor_readings = list(map(int, user_input.split()))

valid_readings = []

for hour , value in enumerate (sensor_readings):
    if value % 2 == 0 :
        valid_readings.append ((hour, value))

print("valid sensor readings (hour, value): ")
print (valid_readings )
```

### Problem Statement 4: Email Domain Usage Analysis


```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]

domain_count = {}

for email in emails:
    domain = email.split("@")[1]

    if domain in domain_count :
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

total_users = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total_users)*100
    print(f"{domain}: {int (percentage)}%")
    
```

### Problem Statement 5: Sales Spike Detection

```python
user_input = input ("enter daily sales seperated by space : ")
sales = list(map(int , user_input.split()))

avg_sales = sum(sales)/ len(sales)
threshold = avg_sales *1.3

for day , value in enumerate (sales, start=1):
    if value> threshold :
        print(f"Day {day}: {value}")
```

### Problem Statement 6: Duplicate User ID Detection

```python
user_in = input ("enter user id seperated by space :")

user_ids = user_in.split()

count_dict = {}

for user in user_ids:
    count_dict[user]= count_dict.get(user, 0)+1

for user, count in count_dict.items():
    if count>1:
        print (f"{user} -> {count}times") 
```
