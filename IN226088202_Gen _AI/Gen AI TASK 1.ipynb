1. Difference between "Love" and "love":
"Love" and "love" are treated as different tokens due to case sensitivity. Converting to lowercase ensures consistency.

2. What happens if stopwords are not removed:
The dataset becomes noisy, increases computational cost, and reduces model accuracy.

3. When removing stopwords is harmful:
- Sentiment analysis (e.g., "not good" loses meaning)
- Legal/medical text where every word matters

4. Stemming vs Lemmatization:
- Stemming: cuts words roughly (running → runn)
- Lemmatization: returns meaningful base form (running → run)

```python
import re
from collections import Counter
import numpy as np
```

### TASK 2 ###

```python
def preprocess_text(text):
    if not isinstance(text, str) or text.strip() == "":
        return [], ""
    
    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)
    
    # Remove emails
    text = re.sub(r"\S+@\S+", "", text)
    
    # Remove numbers
    text = re.sub(r"\d+", "", text)
    
    # Lowercase
    text = text.lower()
    
    # Handle repeated characters (soooo → soo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    
    # Remove special characters & emojis
    text = re.sub(r'[^a-z\s]', '', text)
    
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # Tokenization
    tokens = text.split()
    
    # Remove short tokens (<=2) except "no", "not"
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]
    
    cleaned_sentence = " ".join(tokens)
    
    return tokens, cleaned_sentence
```

### TASK 3 ###

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
```

```python
all_tokens = []

for sentence in test_sentences:
    tokens, clean_sentence = preprocess_text(sentence)
    all_tokens.extend(tokens)
    
    print("Original:", sentence)
    print("Tokens:", tokens)
    print("Cleaned Sentence:", clean_sentence)
    print("-" * 50)
```

### TASK 4 ###

```python
def token_analysis(tokens):
    if len(tokens) == 0:
        return 0, 0, 0
    
    total = len(tokens)
    unique = len(set(tokens))
    avg_len = np.mean([len(t) for t in tokens])
    
    return total, unique, avg_len
```

```python
for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    total, unique, avg_len = token_analysis(tokens)
    
    print(f"Sentence: {sentence}")
    print(f"Total Tokens: {total}")
    print(f"Unique Tokens: {unique}")
    print(f"Avg Token Length: {avg_len:.2f}")
    print("-" * 50)
```

Analysis:

1. Sentence with most noise:
"Win $$$ now!!! Limited offer!!!" contains symbols, repeated punctuation, and noise.

2. Sentence with most meaningful tokens:
"I absolutely loved this product" retains meaningful words after cleaning.

### TASK 6 ###


```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []
    
    for text in text_list:
        tokens, clean_text = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean_text)
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
result = full_pipeline(test_sentences)
print(result)
```
