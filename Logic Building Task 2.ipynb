### Unique words in a sentence 

```python
sentence = input ("Enter sentence : ")
words = sentence.split()
unique_words = set(words)

print("Unique words: ", unique_words)
print("count of unique words : ", len(unique_words))
```

### Highest Salary from Employee Data

```python
employees = {
    "Rvai"   : 75000,
    "Anitha" : 68000,
    "Kiran"  : 72000
}

highest_employee = max (employees, key = employees.get)

print("Highest salary : ", highest_employee, "-", employees[highest_employee])
```

### Problem 3: Find Maximum and Minimum Values


```python
numbers = list(map(int,input("Enter numbers seperated by space: ").split()))

maxi= numbers[0]
mini = numbers[0]

for number in numbers :
    if number > maxi :
        maxi = number 
    if number < mini:
        mini = number

print("List: ", numbers)
print("maximum : ", maxi)
print("minimum : ", mini)
```

### Problem 4: Count Products Above a Price Threshold

```python
products = list(map(int, input( " Enter product prices seperated by space : ").split()))
count =0

for prod in products :
    if(prod > 1000):
        count+=1

print("product prices : ", products )
print("products above 1000 : ", count)
```

### Problem 5: Calculate Attendance Percentage


```python
attendance = input ("enter attendance (P for present and A for absent) seperated by space : ").split()

present_count= 0

for status in attendance :
    if status == "P":
        present_count +=1

total_days = len(attendance)
percentage = (present_count/total_days)*100

print("Attendance Percentage: ", percentage)
```

### Problem 7: Count Character Frequency


```python
text = input("enter a string :")

char_count = {}

for ch in text:
    char_count[ch] = char_count.get(ch, 0)+1
print(char_count)
```

### Problem 9: Check if a Key Exists in a Dictionary

### Problem 10: Calculate Average Marks

```python
marks = list(map(int, input("Enter student marks (seperated by space): ").split()))
summ=0
for mark in marks :
    summ= sum(marks)

avg = summ/len(marks)

print("Average marks is : ", avg)
    
```
